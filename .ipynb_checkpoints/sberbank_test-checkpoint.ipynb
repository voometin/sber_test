### Определяем ряд вспомогательных функций и классов - Preprocessor (применяет к исходному множество признаков OneHotEncoding, LabelEncoding, PowerTransformation, StandardScaling в соответствии с указным множеством признаков для трансформации), cross_validation (производит кросс-валидацию на множестве указанных признаках), hyperparameters_optimization (поиск наиболее оптимальных гиперпараметров в указанном пространстве поиска, улучшающий скор модели во время кросс-валидации), plot_fe (функция для отрисовки важности признаков)

```python
class Preprocessor:#(BaseEstimator, TransformerMixin):
    def __init__(self, ohe_features=[], label_encoding_features=[], power_transform_columns=[], standard_scaling_columns=[], model_type='non-tree'):
        self.ohe_features = ohe_features
        self.label_encoding_features = label_encoding_features
        self.power_transform_columns = power_transform_columns
        self.standard_scaling_columns = standard_scaling_columns
        self.feature_names = []
            
        self.model_type = model_type 
        
        self.ohe = None
        self.le = []
        self.power_transformer = None
        self.standard_scaler = None
        
    def fit(self, df, y=None, copy=True):
        df = df.copy()
            
        if self.ohe_features:
            # OneHotEncoder
            self.ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
            self.ohe.fit(df[self.ohe_features].values)
            # save OneHotEncoder
            with open(f'./preprocessors/one-hot-encoder-{self.model_type}', 'wb') as f:
                pickle.dump(self.ohe, f)

        self.le = []
        for column in self.label_encoding_features:
            # LabelEncoder()
            self.le.append(LabelEncoder())
            self.le[-1].fit(df[column].values.reshape(-1, 1))
            # save LabelEncoder
            with open(f'./preprocessors/label-encoder-{column}-{self.model_type}', 'wb') as f:
                pickle.dump(self.le[-1], f)

        if self.power_transform_columns:
            self.power_transformer = PowerTransformer(standardize=False)
            df[self.power_transform_columns] = self.power_transformer.fit_transform(df[self.power_transform_columns].values)
            # save PowerTransformer
            with open(f'./preprocessors/power-transformer-{self.model_type}', 'wb') as f:
                pickle.dump(self.power_transformer, f)

        if self.standard_scaling_columns:
            self.standard_scaler = StandardScaler()
            self.standard_scaler.fit(df[self.standard_scaling_columns].values)
            # save StandardScaler
            with open(f'./preprocessors/standard-scaler-{self.model_type}', 'wb') as f:
                pickle.dump(self.standard_scaler, f)
                
        return self
    
    def transform(self, df):
        df = df.copy()
        if self.model_type == 'logreg':
            pass
            df['f7_is_null'] = (df.f7.values == 0) * 1
            if 'f7' in self.standard_scaling_columns:
                df.loc[df.f7.values == 0, 'f7'] = df.loc[df.f7.values != 0, 'f7'].mean()


        if self.ohe_features:
            if self.ohe is None:
                with open(f'./preprocessors/one-hot-encoder-{self.model_type}', 'rb') as f:
                    self.ohe = pickle.load(f)
            ohe_features = self.ohe.transform(df[self.ohe_features].values)#.astype(int)
            ohe_feature_labels = [f'{self.ohe_features[index]}_{str(val)}' for index, category in enumerate(self.ohe.categories_) for val in category]
            df.drop(columns=self.ohe_features, axis=1, inplace=True, errors='ignore')
            df = pd.concat([df, pd.DataFrame(ohe_features, index=df.index, columns=ohe_feature_labels)], axis=1)
            
        if self.label_encoding_features:
            if not self.le:
                for column in self.label_encoding_features:
                    with open(f'./preprocessors/label-encoder-{column}-{self.model_type}', 'rb') as f:
                        self.le.append(pickle.load(f))
            for index, column in enumerate(self.label_encoding_features):
                df[column] = self.le[index].transform(df[column].values.reshape(-1, 1))#.astype(int)
            
        if self.power_transform_columns:
            if self.power_transformer is None:
                with open(f'./preprocessors/power-transformer-{self.model_type}', 'rb') as f:
                    self.power_transformer = pickle.load(f)
            df[self.power_transform_columns] = self.power_transformer.transform(df[self.power_transform_columns].values)
            
        if self.standard_scaling_columns:
            if self.standard_scaler is None:
                with open(f'./preprocessors/standard-scaler-{self.model_type}', 'rb') as f:
                    self.standard_scaler = pickle.load(f)
            df[self.standard_scaling_columns] = self.standard_scaler.transform(df[self.standard_scaling_columns].values)
            
        if not self.feature_names:
            self.feature_names = df.columns.tolist()
            
        return df
    
        
            
```

```python
def cross_validation(cv, model, X, y, metrics=[f1_score], verbose=True):
    
    scores = {}
    for metric in metrics:
        scores[metric.__name__] = {'train': [], 'val': []}
    
    for train_index, val_index in cv.split(X, y):
        X_train, X_val, y_train, y_val = X.loc[train_index], X.loc[val_index], y[train_index], y[val_index]

        model.fit(X_train, y_train)
        
        train_predictions_proba = model.predict_proba(X_train).T[1]
        val_predictions_proba = model.predict_proba(X_val).T[1]

        train_predictions = np.round(train_predictions_proba)
        val_predictions = np.round(val_predictions_proba)

        # metric calculation
        for index, metric in enumerate(metrics):
            if metric.__name__ in ['precision_recall_curve', 'roc_curve']:
                train_score = auc(*metric(y_train, train_predictions_proba)[:2][::-1])
                val_score = auc(*metric(y_val, val_predictions_proba)[:2][::-1])
            else:
                train_score = metric(y_train, train_predictions)
                val_score = metric(y_val, val_predictions)

            scores[metric.__name__]['train'].append(train_score)
            scores[metric.__name__]['val'].append(val_score)
            
    for metric in metrics:
        for key in ['train', 'val']:
            scores[metric.__name__][f'{key}_mean'] = np.mean(scores[metric.__name__][key]).round(5)
            scores[metric.__name__][key] = np.round(scores[metric.__name__][key], 5)
        if verbose:
            print(metric.__name__)
            print(f"Train: {scores[metric.__name__]['train']}, mean: {scores[metric.__name__]['train_mean']}")
            print(f"Val: {scores[metric.__name__]['val']}, mean: {scores[metric.__name__]['val_mean']}\n")
    
    return scores
    
```

```python
def plot_fe(series, xlabel='', ylabel='', title='', figsize=(13, 10), font_size=14, sort=True):
    if sort:
        series = series.sort_values(ascending=False)
        
    plt.figure(figsize=figsize);
    sns.set(font_scale=font_size/10.0)
    ax = sns.barplot(x=series.values.reshape(1, -1)[0], y=series.index);
    if xlabel:
        ax.set_xlabel(xlabel, fontsize=font_size)
    if ylabel:
        ax.set_ylabel(ylabel, fontsize=font_size)
    if title:
        ax.set_title(title, fontsize=font_size)
    plt.show()
```

```python
def hyperparameters_optimization(space_search, max_evals, loss=''):

    def objective(space_search):
        pipeline['clf'].set_params(**space_search)
        scores = cross_validation(cv, pipeline, X, y, metrics, verbose=False);
        if loss == 'overfit':
            return {'loss': -scores['f1_score']['val_mean'] + max(0, (scores['f1_score']['train_mean'] - scores['f1_score']['val_mean'])), 
                    'status': STATUS_OK, 'scores': scores, 'params': space_search}
        return {'loss': -scores['f1_score']['val_mean'], 'status': STATUS_OK, 'scores': scores, 'params': space_search}
    
    trials = Trials()
    best = fmin(fn=objective,
                space=space_search,
                algo=tpe.suggest,
                max_evals=max_evals,
                trials=trials)
    
    return best, sorted(trials.results, key=lambda x: x['loss'])
```

## Выбор модели:
Так как у нас обычные табличные данные, где признаки в меньшей степени зависят друг от друга, то в первую очередь стоит начать с простой линейной модели - Логистической регрессии. Если значение таргета зависит от комбинации признаков, а зависимость нелинейная, то можно попробывать решающие деревья и дальше попробовать более сложные модели, такие как решающие деревья и градиентный бустинг, feed-forward NN... 

Также на выбор модели должны сказываться ограничения с точки зрения памяти и скорости предсказания (с учетом времени, трубующейся на предобработку признаков). Так, например логистическая регрессия более требовательна к предобработке признаков (приведение признаков к одноме масштабу и обработка категориальных признаков), что влияет на скорость инференса, однако с точки зрения памяти - она необходима, только для хранения моделей по обработке признаков и хранения вектора весов. Модели основанные на деревьях решений более требовательны к памяти, для хранения всех правил в листах деревьев, однако менее чувствительны для обработки признаков, что сказывается на время инференса в лучшую сторону. Одно дерево достаточно быстро в предсказании. Случайный лес и градиентный бустинг требует значительно больше памяти для хранения всех деревьев, но из-за большей сложности моделей, они могут построить более точную разделяющую плоскость.

Так как мне не поступило, каких либо явных ограничений на используемые ресурсы, то попробуем сами оценить исходя из возможных case-ов использования данной модели:

1) Данная модель нужна просто для аналитики - для выявления причин, почему происходит отток клиентов? От каких параметров это зависит? Ответы на эти вопросы нужны для изменения политики банка по отношению к клиентам, которые более склонны к оттоку, с целью их удержания.

2) Данная модель нужна, для дальнейшего формирования некоторого персонализированного предложения с целью удержания клиента склонного к оттоку.

....

В первом случае, нам важна интерпретируемость модели и точность предсказаний. Время инференса и объем потребляемых ресурсов носит второстепенный характер, так как подразумевается, что модель будет использоваться единично.

Во втором случае нам более важна точность предсказаний. В случае, если таргет в большей степени зависит от частоизменяющихся признаков и предсказания будут использоваться в модели с высокочастотными прогнозами, то нам будет важна скорость инференса, и среди моделей с относительно одинаковыми по точности прогнозами, я бы отдал предпочтение наиболее быстрой модели. В случае, если таргет в большей степени зависит от редкоизменяющихся, либо неизменяющихся признаков (возраст, пол, зарплата, ...), то в данном случае, я бы отдал предпочтение более точной модели, так как предсказания модели будут меньше подвержены изменениям с течением времени. Если у нас есть 1 модель, которая требует много вычислительных ресурсов для предсказания и предсказывает значительно точные резлуьтаты, относительно 2-ой модели, которая требует мало вычеслительных ресурсов, но дает более скудные прогнозы, то вероятно следует провести А/B тест в течении определенного времени и замерить прибыль от этих моделей ("Заработок от моделей" - "затраты на их поддержания") и выбрать ту, которая приносит наибольшую прибыль.

В целом, если мало времени, то в данном случае я бы попробывал Логистическую регрессию и градиентный бустинг. Логистическую регрессию для проверки способности модели восстановить линейную зависимость между таргетом и признаками. А градиентный бустинг, потому что зачастую он дает очень хорошее качество предсказаний, по сравнению с другими моделями для обычных табличных данных (при условии, что у нас достаточно данных, по сравнению с множеством признаков). Так как у нас достаточно мало признаков для предсказания таргета по сравнению с количеством данных, то это еще 1 + в пользу того, чтобы попробывать более сложные нелинейные модели.

Если есть достаточно времени, то я бы не ограничивался в данном случае логистической регрессией и градиентным бустингом, а попробовал бы еще другие модели - деревья решений, knn, случайный лес, нейроннные сети.

## Выбор метрики:

Выбор метрики в данном случае зависит от применения модели (case-а) и от распределения таргета в нашей выборке. Исходя из распределения таргета можно сделать выбор, что наша выборка несбалансирована. В качестве метрик для оценки моделей на несбалансированной выборке можно использовать precision, recall, f1, f1_beta, precision_recall_curve_auc. 

Для 1-ого case-а нам важно точное предсказание 1-ого класса, и как следствие нам важен recall, с целью формирования профиля клиента, подверженного оттоку. Однако если модель будет показывать хороший recall и плохой precision, то составленный профиль клиента, подверженного оттоку на основе наших предсказаний, будет сильно зашумлен. Для соблюдения баланса между recall и precision можно использовать в качестве метрики f1, если отдается предпочтение больше одной из этих характеристик, то можно использовать f1_beta. Также, хорошей мерой качества классификатора может служить precision_recall_curve_auc.

Для 2-ого case-а, если мы отдадим предпочтение recall и меньше будем обращать внимание на precision, то мы рискуем получить меньшую прибыль (так как мы будем предлагать в конечном результате например льготную кредитную ставку клиенту, который от нас и не собирался уходить). Если мы будет отдавать предпочтение precision, тогда мы рискуем потерять клиентов (за счет ухудшения recall), которых можно было сохранить путем например предоставления той же льготной кредитной ставке. В данном случае также можно использовать f1, f1_beta, либо precision_recall_curve_auc.

При подборе гиперпараметров я ориентировался на f1 метрику, однако при кросс-валидации оцениваются и другие вышеуказанные метрики.

## Валидация

Так как у нас несбалансированная выборка, то при разбиении на валидационную и тренировчную выборки мы должны учитывать распределение классов в каждой подвыборке. Поэтому в данном случае следует использовать стратифицированную версию валидации по таргету. Среди множества стратифицированных способов валидации я выбрал K-fold валидацию, чтобы была возможность оценить качество моделей на всем наборе данных и сравнительно небольшой объем выборки позволял нам валидироваться достаточно быстро.

## Интерпретация модели.

С увеличением сложности модели усложняется и интерпретация модели. Однако есть ряд библиотек, позволяющие интерпретировать важность признаков как в целом в самой модели, так и для объяснения прогноза модели по заданным входным признакам. Интерпретация сложных моделей происходит за счет использования более простых и интерпретируемых моделей. Для отображения важности признаков использовалась библиотека eli5. Для объяснения отдельных прогнозов можно также использовать eli5, либо lime, либо shap (наиболее популярные), либо любую другую библиотеку. Для объяснения влияния отдельных признаков на прогноз модели в изоляции от других признаков в целом по датасету использовалась библиотека skater.

P.S.: В ходе построения моделей ниже можно заметить, что при предсказании используются не все признаки. Это обуславливается тем, что ниже уже представлены субоптимальные гиперпараметры и ряд признаков был отброшен исходя из их бесполезности для модели и улучшения скора на валидации (не везде - например для решающих деревьев был оставлен признак deposit, которые не использовался, но был оставлен для наглядности и доказательства его бесполезности при отображения важности признаков).
Часть кода, отвечающая за подбор гиперпараметров закоментирована - если необходимо, то раскоментируйте ее.

## Ниже описаны функции для отображения влияния признаков в изоляции от других на итоговый прогноз

```python
def local_plot_partial_dependecies(X):
    ncols = 4
    nrows = ceil(X.shape[1] / 2 / ncols)

    fig, axarr = plt.subplots(nrows, ncols, figsize=(16, 10))
    fig.tight_layout()
    
    columns = X.columns.tolist()[::2]
    for i in range(len(columns)):
        axes = axarr[i//ncols][i%ncols]
        axes.set_title(columns[i])
        if (~X[columns[i]].isna()).sum()>5:
            # continous
            index = ~X[columns[i]].isna()
            axes = sns.lineplot(X.loc[index, columns[i]].values, X.loc[index, columns[i]+'_y'].values, ax=axes)
        else:
            # category/binary
            axes = sns.barplot(X.loc[index, columns[i]].values, X.loc[index, columns[i]+'_y'].values, ax=axes)
        axes.set_ylim(0, 1)

class InverseTransformer():
    
    def __init__(self, preprocessor):
        self.preprocessor = preprocessor
        
    def transform(self, X):
        X = X.copy()
        columns = X.columns
        
        if self.preprocessor.standard_scaling_columns and self.preprocessor.standard_scaling_columns[0] in X.columns:
            X[self.preprocessor.standard_scaling_columns] = self.preprocessor.standard_scaler.\
                                                                inverse_transform(X[self.preprocessor.standard_scaling_columns])
            
        if self.preprocessor.power_transform_columns and self.preprocessor.power_transform_columns[0] in X.columns:
            X[self.preprocessor.power_transform_columns] = self.preprocessor.power_transformer.\
                                                                inverse_transform(X[self.preprocessor.power_transform_columns])
            
        if self.preprocessor.label_encoding_features and self.preprocessor.label_encoding_features[0] in X.columns:
            for index, column in enumerate(self.preprocessor.label_encoding_features):
                X.loc[~X[column].isna().values, column] = self.preprocessor.le[index].\
                                inverse_transform(X.loc[~X[column].isna().values, column].astype(int))
                
        return X
    
def plot_partial_dependencies(features, model, feature_names):
    interpreter = Interpretation(features, feature_names=feature_names)
    pyint_model = InMemoryModel(model['clf'].predict_proba, examples=features, target_names=['p(Churn=0)', 'p(Churn=1)'])
    local_df = pd.DataFrame()

    for col in feature_names:
        tmp = interpreter.partial_dependence.partial_dependence([col], pyint_model, progressbar=False);
        local_df = pd.concat((local_df, pd.DataFrame(tmp.loc[:, [col, 'p(Churn=1)']].values, columns=[col, f'{col}_y'])), axis=1)

    inv_transformer = InverseTransformer(model['preproc'])
    local_df = inv_transformer.transform(local_df)
    local_plot_partial_dependecies(local_df);
```

### KNN

### Так как в классе KNNClassifier нет параметра по взвешиванию отдельных точек в соответствии с весами классов ({class_weight: 'balanced'}), пришлось написать свой класс ClassWeightBalancedKNNClassifier (относительно обычного KNNClassifier класса, дает значительный прирост в f1 метрике). Класс LogRegWeight взвешивает признаки в соответствии с их линейной важности по отношению к таргету. При использовании KNN возникает сложности в интерпретации модели.

```python
class LogRegWeight(BaseEstimator, TransformerMixin):
    
    def __init__(self, params):
        self.model = LogisticRegression(**params)
        
    def fit(self, X, y):
        self.model.fit(X, y)
        return self
    
    def transform(self, X):
        return X * abs(self.model.coef_[0])
```

```python
class ClassWeightBalancedKNNClassifier(BaseEstimator):
    
    def __init__(self, **knn_params):
#         print(knn_params)
        self.knn_params = knn_params
        self.weights = self.knn_params.pop('weights', False)
    
    def fit(self, X, y=None):
        self.X = X.values
        self.y = y
        self.class_weigts = X.shape[0] / np.array([(2 * (X.shape[0] - y.sum())), (2 * y.sum())])
        self.knn = NearestNeighbors(**self.knn_params)
        self.knn.fit(self.X)
        return self
    
    def predict_proba(self, X):
        X = X.values
        
        preds = []
        for batch_index in np.array_split(np.arange(0, X.shape[0]), max(1, int(X.shape[0]/100))):
            distances, indexes = self.knn.kneighbors(X[batch_index])
            values = self.y[indexes.ravel()].reshape(indexes.shape)
            if self.weights:
                batch_preds = ((values * distances * self.class_weigts[values]).sum(axis=1)
                                / (distances * self.class_weigts[values]).sum(axis=1))
            else:
                batch_preds = (values * self.class_weigts[values]).sum(axis=1) / self.class_weigts[values].sum(axis=1)
            preds = np.concatenate((preds, batch_preds))
            
        return np.array([1 - preds, preds]).T
        
        
    def predict(self, X):
        return np.round(self.predict_proba(X).T[1])
    
    def set_params(self, **knn_params):
        self.knn_params = knn_params
        self.weights = self.knn_params.pop('weights', False)
```

```python
# define X and Y
cols = ['f2', 'f5', 'f6',
       'f7', 'f8', 'f10',
        'f4', 'f9', 'f11'
       ]
X, y = df.loc[:, cols], df.churn.values

# define model
preproc_params = {
                'ohe_features': ['f5', 'f8',],
                'label_encoding_features': ['f6',],
                'power_transform_columns': ['f2', 'f4', ],
                'standard_scaling_columns': ['f2', 'f7', 'f4', 'f11',],
                'model_type': 'knn'
}              
knn_logreg_params = {'C': 0.1, 'class_weight': 'balanced',}
model_params = {'n_neighbors': 188, 'p': 2, 'weights': False}
# {'n_neighbors': 188, 'p': 2, 'weights': False} - {f1_score_train_mean: 0.52783, f1_score_val_mean: 0.52517}
# {'n_neighbors': 146, 'p': 2, 'weights': True} - {f1_score_train_mean: 0.51786, f1_score_val_mean: 0.51987}
# {'n_neighbors': 127, 'p': 1, 'weights': False} - {f1_score_train_mean: 0.52875, f1_score_val_mean: 0.52309}

pipeline = Pipeline([
    ('preproc', Preprocessor(**preproc_params)),
    ('logreg', LogRegWeight(knn_logreg_params)),
    ('clf', ClassWeightBalancedKNNClassifier(**model_params)),
])

# define cross_validation
cv_params = {
    'n_splits': 8,
    'random_state': 8732,
    'shuffle': True,
}
cv = StratifiedKFold(**cv_params)

# define metrics to calculate during cross_validation
metrics = [precision_score, recall_score, f1_score, precision_recall_curve, accuracy_score]

# hyperparameters tuning
# search_space = {'n_neighbors': hp.uniformint('n_neighbors', 5, 300),
#                'weights': False,
#                'p': 2,
#                }
# max_eval = 20
# best_params, hp_tuning_results = hyperparameters_optimization(search_space, max_eval)

# best_params = hp_tuning_results[0]['params']
# print(best_params)

# pipeline['clf'].set_params(**best_params)

# fit and validate the model according to given CV strategy and given metrics
cross_validation(cv, pipeline, X, y, metrics);

pipeline.fit(X, y)
# measure inference time on all dataset
tm = time.time()
# predictions['knn'] = pipeline.predict_proba(X).T[1]
pipeline.predict_proba(X).T[1]
print(f'INFERENCE TIME FOR ALL DATASET: {time.time() - tm}')

with open('./models/knn', 'wb') as f:
    pickle.dump(pipeline, f)
```

### Stacking

```python
class ColumnsSelector(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns):
        self.columns = columns
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X[self.columns].copy()
    
class ModelTransformerWraper(BaseEstimator, TransformerMixin):

    def __init__(self, model):
        self.model = model
        
    def fit(self, X, y=None):
        self.model.fit(X, y)
        return self
    
    def transform(self, X):
        return self.model.predict_proba(X).T[1].reshape(-1, 1)
```

```python
# logreg
logreg_cols = ['f2', 'f5', 'f6',
       'f7', 'f8', 'f10',
       ]

logreg_preproc_params = {
                'ohe_features': ['f5', 'f8',],
                'label_encoding_features': ['f6',],
                'power_transform_columns': ['f2',],
                'standard_scaling_columns': ['f2', 'f7',],
                'model_type': 'non-tree'
}              
logreg_params = {'C': 0.1, 'class_weight': 'balanced',}

# decision_treee
dt_cols = ['f2', 'f3', 'f4', 'f5', 'f6',
       'f7', 'f8', 'f9', 'f10',
         'f11'
       ]

dt_preproc_params = {
                'ohe_features': [],
                'label_encoding_features': ['f5', 'f6'],
                'power_transform_columns': [],
                'standard_scaling_columns': [],
                'model_type': 'tree'
}             
dt_params = {'max_depth': 5, 'class_weight': 'balanced',}

# knn
knn_cols = ['f2', 'f5', 'f6',
       'f7', 'f8', 'f10',
         'f4', 'f9', 'f11'
       ]
knn_preproc_params = {
                'ohe_features': ['f5', 'f8',],
                'label_encoding_features': ['f6',],
                'power_transform_columns': ['f2', 'f4', ],
                'standard_scaling_columns': ['f2', 'f7', 'f4', 'f11',],
                'model_type': 'non-tree'
}              
knn_params = {'n_neighbors': 188, 'p': 2, 'weights': False}
# {'n_neighbors': 188, 'p': 2, 'weights': False} - {f1_score_train_mean: 0.52783, f1_score_val_mean: 0.52517}
# {'n_neighbors': 146, 'p': 2, 'weights': True} - {f1_score_train_mean: 0.51786, f1_score_val_mean: 0.51987}
# {'n_neighbors': 127, 'p': 1, 'weights': False} - {f1_score_train_mean: 0.52875, f1_score_val_mean: 0.52309}

# rf
rf_cols = ['f2', 'f3', 'f4', 'f5', 'f6',
       'f7', 'f8', 'f9', 'f10', 'f11'
       ]

rf_preproc_params = {
                'ohe_features': [],
                'label_encoding_features': ['f5', 'f6'],
                'power_transform_columns': [],
                'standard_scaling_columns': [],
                'model_type': 'tree'
}             
rf_params = {'n_estimators': 150, 'max_depth': 5, 'class_weight': 'balanced', 'random_state': 9273}

# lgb
lgb_cols = ['f2', 'f5', #'f6',
       'f7', 'f8', 'f10',
       ]

lgb_preproc_params = {
                'ohe_features': [],
                'label_encoding_features': ['f5'],
                'power_transform_columns': [],
                'standard_scaling_columns': [],
                'model_type': 'tree'
}             
lgb_params = {'class_weight': 'balanced', 'n_estimators': 16, 'num_leaves': 7, 'reg_lambda': 2.4282700956821612}

feature_concat = FeatureUnion(transformer_list=[
        ('logreg', Pipeline([
            ('ColumnsSelector', ColumnsSelector(logreg_cols)),
            ('preproc', Preprocessor(**logreg_preproc_params)),
            ('clf', ModelTransformerWraper(LogisticRegression(**logreg_params))),
        ])),
        ('dt', Pipeline([
            ('ColumnsSelector', ColumnsSelector(dt_cols)),
            ('preproc', Preprocessor(**dt_preproc_params)),
            ('clf', ModelTransformerWraper(DecisionTreeClassifier(**dt_params))),
        ])),
        ('knn', Pipeline([
            ('ColumnsSelector', ColumnsSelector(knn_cols)),
            ('preproc', Preprocessor(**knn_preproc_params)),
            ('knn_logreg', LogRegWeight(logreg_params)),
            ('clf', ModelTransformerWraper(ClassWeightBalancedKNNClassifier(**knn_params))),
        ])),
        ('rf', Pipeline([
            ('ColumnsSelector', ColumnsSelector(rf_cols)),
            ('preproc', Preprocessor(**rf_preproc_params)),
            ('clf', ModelTransformerWraper(RandomForestClassifier(**rf_params))),
        ])),
        ('lgb', Pipeline([
            ('ColumnsSelector', ColumnsSelector(lgb_cols)),
            ('preproc', Preprocessor(**lgb_preproc_params)),
            ('clf', ModelTransformerWraper(LGBMClassifier(**lgb_params))),
        ])),
])
feature_concat_columns = [i[0] for i in feature_concat.transformer_list]
cv_params = {
    'n_splits': 8,
    'random_state': 8732,
    'shuffle': False,
}
cv = StratifiedKFold(**cv_params)

level_0_predictions = pd.DataFrame(index=df.index, columns=feature_concat_columns)

for train_index, val_index in cv.split(df, y):
        X_train, X_val, y_train = df.loc[train_index], df.loc[val_index], y[train_index]

        feature_concat.fit(X_train, y_train)
        
        level_0_predictions.loc[val_index, feature_concat_columns] = feature_concat.transform(X_val)

level_0_predictions = level_0_predictions.astype(float)
# define metrics to calculate during cross_validation
metrics = [precision_score, recall_score, f1_score, precision_recall_curve, accuracy_score]

logreg = LogisticRegression(**{'class_weight': 'balanced'})
# fit and validate the model according to given CV strategy and given metrics
cross_validation(cv, logreg, level_0_predictions, df.churn.values, metrics);

level_0_predictions['churn'] = df.churn.values
```
